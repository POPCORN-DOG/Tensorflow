# 최소제곱법(Least Square Method)

최소제곱법은 실제 딥러닝을 실행할 때 만나는 문제에서 대부분 활용할 수 없다.<br>
그 이유는 최소제곱법을 계산하기 위해 꼭 필요한 조건들을 알 수 없기 때문이다.<br>
따라서 미분과 기울기를 이용해야 한다.<br><br>

딥러닝은 결국 일차 함수의 a(기울기),b(절편)값을 구하는 것인데, <br>
a와 b값은 이차 함수 포물선의 최솟값을 미분으로 구할 수 있다.<br>
최솟값에 도달하기 위해서는 점과 포물선이 접한 접선의 기울기가 0이 되도록 찾아가야한다.<br>
그 접선의 기울기를 구하기 위한 장치가 바로 미분이다. 


- 최소 제곱법은 주어진 x의 값이 하나일 때 적용이 가능하다. 
- 여러 개의 x가 주어지는 경우는 경사 하강법을 이용한다. 
- 여기서의 입력값이 x(공부한 시간) y(성적)일 떄 기울기 a를 구하는 방법은 다음과 같다.


```python
import numpy as np
```

```python
x = [2, 4, 6, 8]
y = [81, 93, 91, 97]
```

```python
# x와 y의 평균값
mx = np.mean(x)
my = np.mean(y)
```

```python
# 기울기 공식의 분모
divisor = sum((i - mx)**2 for i in x)
```
