
- 독립적으로 변할 수 있는 x값 독립 변수와 이에 따라 종속적으로 변하는 y값 종속 변수<br>
- 선형 회귀란 독립변수 x를 사용해 종속 변수 y의 움직임을 예측하고 설명하는 작업.<br>
- `단순 선형 회귀(simple linear regression)`: 하나의 x값만으로도 y값을 설명할 수 있을 때 <br>
- `다중 선형 회귀(multiple linear regression)`: 여러개의 x값으로 y값을 설명할 수 있을 때 <br>


```python
# 딥러닝에 필요한 케라스 함수
# 딥러닝을 여행이라고 표현하면 
# 텐서플로는 목적지까지 빠르게 이동시켜주는 비행기
# 케라스는 비행기의 이륙 및 정확한 지점까지의 도착을 책임지는 파일럿
from keras.models import Sequential
from keras.layers import Dense

# 필요한 라이브러리를 불러온다.
# 라이브러리란 특정 기능을 담은 작은 프로그램(모듈, module)을 말한다.
# 즉, 함수나 클래스를 따로 담아 라이브러리 형태로 만든다.
import numpy as np
import tensorflow as tf
```

```python
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 폴란드의 브로츠와프 의과대학에서 2013년 공개한 폐암 수술 환자의 
# 수술 전 진단 데이터와 수술 후 생존 결과를 기록한 실제 의료 기록 데이터이다.
# 속성(attribute) 1:17까지 종양의 유형, 폐활량, 호흡 곤란 여부 등
# 클래스(class) 1 수술 후 생존, 0 수술 후 사망
data_set = np.loadtxt("dataset/ThoraricSurgery.csv",delimiter=",")
print(data_set.shape)

# import pandas as pd
# pd.DataFrame(data_set[:,17])
# [시작행:끝행 , 시작열:끝열]
x = data_set[:,0:17]
y = data_set[:,17]
```

```python
# Sequential 함수의 model.add()함수를 사용해 층을 차례로 추가한다.
# 층을 몇 개 쌓을지는 데이터에 따라 결정한다. 
# 케라스의 큰 장점 중 하나는 model.add() 함수를 이용해 필요한 만큼의 층을 빠르게 쉽게 쌓아올릴 수 있다는 것.

# dense() 함수 '조밀하게 모여있는 집합'이란 뜻으로, 각 층의 특성 옵션을 설정한다.
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 딥러닝의 구조와 층별 옵션을 정하고 나면 compile() 함수를 이용해 실행한다.
# loss: 한 번 신경망이 실행될 때마다 오차 값을 추적하는 함수
# optimizer: 오차를 어떻게 줄여 나갈지 정하는 함수.
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(x,y, epochs=100, batch_size=10)
```

```python
# model.evaluate()함수로 모델의 정확도를 점검
# 여기서의 정확도는 학습 대상이 되는 기존 환자들의 데이터 중에 일부를 랜덤하게 추출해,
# 새 환자인 것으로 가정하고 테스트한 결과이다.

# 좀 더 신뢰할 수 있는 정확도를 측정하려면, 학습 단계에서 미리 테스트셋을 이용하여 측정하자.
print("\n Accuracy: %.4f" % (model.evaluate(x, y )[1]))
```
